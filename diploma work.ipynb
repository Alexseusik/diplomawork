```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'dataset_2024-05-06T14_22_32.523577261Z_DEFAULT_INTEGRATION_SSSU_DF_PRICE_CHANGE_CONSUMER_GOODS_SERVICE_21.0.0.csv'
```

```python
df = pd.read_csv(file)
```

```python
prices_dataset = df[df['Показник'] == 'Середні споживчі ціни на товари (послуги)'].reset_index()
```

```python
prices_dataset
```

```python
columns_to_drop = [
    'index',
    'Масштабування', 
    'Кількість десяткових знаків', 
    'Конфіденційність', 
    'Примітки часового ряду', 
    'Примітки спостереження',
    'Базисний період', 
    'Показник',
    'Періодичність'
]

prices_dataset = prices_dataset.drop(columns=columns_to_drop)
prices_dataset
```

```python
date_columns = prices_dataset.columns[prices_dataset.columns.str.contains(r'^\d{4}-M\d{2}$')]
prices_dataset = prices_dataset.rename(columns={col: col.replace('M', '') for col in date_columns})
```

```python
prices_dataset
```

```python
empty_columns = [col for col in prices_dataset.columns if prices_dataset[col].isna().all()]
prices_dataset = prices_dataset.drop(columns=empty_columns)
```

```python
prices_dataset
```

```python
apples_prices = prices_dataset[(prices_dataset['Тип товарів і послуг'] == 'Яблука') & 
                                       (prices_dataset['Територіальний розріз'] == 'Україна')].reset_index(drop=True)
```

```python
apples_prices
```

Подивимось на початковий аналіз даних. Для початку переведемо нашу таблицю в робочий варіант. Для цього транспонуємо нашу таблицю і виведемо лише значення ціни. Подивимось на середнє значення, дисперсію і 

```python
date_columns = apples_prices.columns[3:]
prices = apples_prices[date_columns].transpose()
prices.index = pd.to_datetime(prices.index, format='%Y-%m')
prices = prices.reset_index()
prices.columns = ['Date', 'Price']


prices
```

```python
mean_apple = np.mean(prices['Price'])
var_apple = np.var(prices['Price'])
sd_apple = np.sqrt(var_apple)
max_price = np.max(prices['Price'])
min_price = np.min(prices['Price'])

mean_apple, var_apple, sd_apple, max_price, min_price
```

```python
Q1 = prices['Price'].quantile(0.25)
Q3 = prices['Price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = prices[(prices['Price'] < lower_bound) | (prices['Price'] > upper_bound)]

outliers.sort_values(by='Price')
```

```python
plt.figure(figsize=(12, 8))
plt.boxplot(prices['Price'].dropna(), vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='black'),
            whiskerprops=dict(color='black'),
            capprops=dict(color='black'),
            medianprops=dict(color='red'),
            flierprops=dict(marker='o', color='red', alpha=0.5))

plt.title('Ціни на яблука')
plt.xlabel('Ціна')
plt.grid(True)

# Annotate outliers with dates and values: two up-left, one down-right
for i, (_, row) in enumerate(outliers.sort_values(by='Price').iterrows()):
    if i % 2 == 0:
        y_offset = -20, 
    else:
        y_offset = 20
    plt.annotate(f"{row['Date'].strftime('%Y-%m')}, {row['Price']}", 
                 (row['Price'], 1), textcoords="offset points", 
                 xytext=(0 , y_offset), ha='center', color='red', 
                 fontsize=8, arrowprops=dict(arrowstyle="->", color='black'))

plt.show()
```
