```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = 'dataset_2024-05-06T14_22_32.523577261Z_DEFAULT_INTEGRATION_SSSU_DF_PRICE_CHANGE_CONSUMER_GOODS_SERVICE_21.0.0.csv'
```

```python
df = pd.read_csv(file)
```

```python
prices_dataset = df[df['Показник'] == 'Середні споживчі ціни на товари (послуги)'].reset_index()
```

```python
prices_dataset
```

```python
columns_to_drop = [
    'index',
    'Масштабування', 
    'Кількість десяткових знаків', 
    'Конфіденційність', 
    'Примітки часового ряду', 
    'Примітки спостереження',
    'Базисний період', 
    'Показник',
    'Періодичність'
]

prices_dataset = prices_dataset.drop(columns=columns_to_drop)
prices_dataset
```

```python
date_columns = prices_dataset.columns[prices_dataset.columns.str.contains(r'^\d{4}-M\d{2}$')]
prices_dataset = prices_dataset.rename(columns={col: col.replace('M', '') for col in date_columns})
```

```python
prices_dataset
```

```python
empty_columns = [col for col in prices_dataset.columns if prices_dataset[col].isna().all()]
prices_dataset = prices_dataset.drop(columns=empty_columns)
```
